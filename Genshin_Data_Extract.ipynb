```python
url4 = 'https://spiralabyss.org'
web_page = bs4.BeautifulSoup(requests.get(url4,{}).text, "lxml")
tables = web_page.find_all(name="table", attrs= {"class": "table table-striped table-dark mb-3 table-no-x"})
table_rows = tables[-1].find_all(name="tr")
len(table_rows)
```

```python
col_names = [raw_name.text.replace("\n","") for raw_name in table_rows[0].find_all(name = "th")]
col_names
```

```python
#Taking as a dictionary
contents = []
for row_id in range(2, len(table_rows)):
  content = {col_names[idx]:raw_name.text.replace("\n","").strip() for idx, raw_name in enumerate(table_rows[row_id].find_all(name = "td"))}
  contents.append(content)
```

```python
contents[0]
```

```python
character_spiral = pd.DataFrame(data=contents, columns=['Name','Used %'])
character_spiral.head()
```

```python
character_spiral.rename(columns = {'Name':'Character','Used %': 'Spiral Uses %'}, inplace = True)
character_spiral.head()
```

```python
character_list = character_list.merge(character_spiral)
character_list.head(10)
```
